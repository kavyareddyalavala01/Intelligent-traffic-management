```python
import numpy as np

# Define class labels for your dataset
class_labels = {
    0: 'Auto_rickshaws', 1: 'ambulance', 2: 'bicycle', 3: 'bus', 4: 'car', 5: 'construction_vehicle', 6: 'mini_truck', 7: 'motorbike', 8: 'pickup_truck', 9: 'tempo_loader', 10: 'tempo_traveller', 11: 'tractor_trolly', 12: 'truck'}

def predict_image_class(model, img_path, class_labels):
    results = model.predict(source=img_path, save=False, conf=0.25)
    if results and len(results[0].boxes) > 0:
        predicted_class_index = int(results[0].boxes.cls[0])
        print(f"Predicted class index: {predicted_class_index}")
        predicted_class_label = class_labels.get(predicted_class_index, "Unknown")
        print(f"Predicted class label: {predicted_class_label}")
    else:
        predicted_class_label = "No object detected"
    return predicted_class_label

```

```python
# Load the trained model for inference
infer = YOLO("/content/runs/detect/train/weights/best.pt")  # Replace with the correct model path

# Path to the input image
image_path = '/content/drive/MyDrive/Try.v3i.yolov8/train/images/101Truck_jpg.rf.302ddb937a7a16ca87fd6893deb9081f.jpg'

# Get the prediction for the image
predicted_class = predict_image_class(infer, image_path, class_labels)

# Print the predicted class
print("Predicted class:", predicted_class)
```

```python
import json

# Inverse dictionary to map class labels back to indices
class_ind = {v: k for k, v in class_labels.items()}

# Save the class indices to a JSON file
with open("class_ind.json", "w") as f:
    json.dump(class_ind, f)
```
